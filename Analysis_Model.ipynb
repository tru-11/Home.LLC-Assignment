```python
#Importing libraries as required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
#Reading the prepared final dataset into a dataframe

house_price_us = pd.read_csv("/content/Final_Dataset _Prepared.csv").set_index("DATE")
house_price_us.head()
```

The Year and Month is column is dropped as it is not required only for data preparation it was added.

```python
# Dropping year and month columns from dataframe

house_price_us.drop(columns = [ "Year", "Month"], inplace=True)
```

```python
house_price_us.drop(columns = [ "Sl"], inplace=True)
```

###Analysis Part

Calculate the correlation matrix to measure the linear relationships between
variables.





```python
#Correlation among the variables in dataframe

house_price_us.corr()
```

```python
# Separating the target variable and the independent variable
t1 = house_price_us.pop("CSUSHPISA")
X = house_price_us
```

```python
# Plotting scatter plots of the CASE-SHILLER index vs features
for feature in X.columns:
    plt.figure()
    plt.scatter(x = X[feature], y=t1)
    plt.xlabel(feature)
    plt.ylabel("CASE-SHILLER index")
    plt.title(f"CASE_SHILLER index vs {feature}")
```

Using the above correlation matrix and scatter plots,, the following conclusions can be made-



*  We can see that the unemployment rate has a slightly negative correlation with home prices.
*   The number of new homes has a correlation of almost 0. This is surprising as one would expect that the supply of new homes would also have an effect on home prices. Because maybe the number of new houses built are not enough to bring the prices down significantly.

*  The impact of the Great Recession is reflected in most of the plots.
*  The interest rates have a slightly negative impact on home prices.

Let’s look at the correlation matrix: Each pair of independent variables has a high correlation to the target.
Let’s take one column out of each of these pairs because they will have a low correlation. We’ll drop the column that has a lower correlation to the target below.

```python
# Dropping multicollinearity columns

mult_cols = ["Working_age_pop", "CPI", "urban_pop_us", "Num_Households", "Subsidy", "old_age_pop", "Cons_material", "median_income"]
house_price_us.drop(columns = mult_cols, inplace=True)
X = house_price_us
```

###Building Model

```python
# Scaling
scalar = MinMaxScaler()
X = scalar.fit_transform(X)
```

```python
# Splitting data into train and validation sets
X_train, X_valid, t1_train, t1_valid = train_test_split(X,t1, test_size= 0.2, random_state= 42)
```

```python
# Linear Regression Model
model = LinearRegression()
model.fit(X_train, t1_train)
pred = model.predict(X_valid)
score = r2_score(pred, t1_valid)
print("The r2_score for the validation set is: ", score)
```

The r2_score for the validation set is:  0.9227378976178473.

 The r2 score is close to 1. It means that the model predicts the target variable with good accuracy.


```python
#Printing the regression coefficients of different variables
coefs = model.coef_
cols = house_price_us.columns
for i in range(len(coefs)):
    print(f"The coefficient for {cols[i]} is {coefs[i]}")
```

```python
# The intercept
print("The intercept is: ", model.intercept_)
```
